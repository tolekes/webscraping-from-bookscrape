```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
All_books = {}
```

```python
pages = [str(i) for i in range(1,51)]
```

```python
for page in pages:
    my_url = 'http://books.toscrape.com/catalogue/page-'+ page +'.html'

    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()


    page_soup = soup(page_html, "html.parser")


    book_containers = page_soup.findAll(attrs = {'class':'product_pod'})


    for book in book_containers:
        #book_name.append(book.h3.a.get('title'))
        #book_price.append(book.find('p',attrs = {'class':'price_color'})).text[1:]
        #book_price.append(book.find('p', class_="price_color").text[1:])
        All_books[book.h3.a.get('title')] = book.find('p', class_="price_color").text[1:]
```

```python
Allbooks = pd.DataFrame(list(All_books.items()), columns = ['Book title','Price ($)'])
```
